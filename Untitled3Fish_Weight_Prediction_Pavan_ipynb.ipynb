# Fish Weight Prediction

## Objective

Predict the weight of a fish using its measurements such as height, width, and length by using Linear Regression.

## Libraries Used

- Pandas
- Scikit-learn

```python
# Step 1: Import library

import pandas as pd
```

```python
# Step 2 : Import data

fish = pd.read_csv(
    'https://github.com/ybifoundation/Dataset/raw/main/Fish.csv'
)
```

```python
fish.head()
```

```python
# Step 4 : Check rows and columns

fish.shape
```

```python
# Step 5 : Display column names

fish.columns
```

```python
# Step 6 : Check missing values

fish.isnull().sum()
```

```python
# Step 7 : Dataset information

fish.info()
```

```python
# Step 8 : Select input columns

X = fish[['Category',
          'Height',
          'Width',
          'Length1',
          'Length2',
          'Length3']]
```

```python
# Step 9 : Select output column

y = fish['Weight']
```

```python
X.head()
```

```python
y.head()
```

```python
# Step 11 : Import train_test_split

from sklearn.model_selection import train_test_split
```

```python
# Step 12 : Split data into training and testing

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=2529
)
```

```python
# Step 13 : Check data sizes

X_train.shape, X_test.shape
```

```python
# Step 14 : Import Linear Regression

from sklearn.linear_model import LinearRegression
```

```python
# Step 15 : Create model

model = LinearRegression()
```

```python
# Step 16 : Train model

model.fit(X_train, y_train)
```

```python
# Step 17 : Predict weight

y_pred = model.predict(X_test)
```

```python
# Step 18 : Display predicted values

y_pred[:5]
```

```python
# Step 19 : Check accuracy score

model.score(X_test, y_test)
```

## Conclusion

The Linear Regression model was trained successfully.

The model was able to predict fish weight using height, width, and length measurements.

The obtained R² score was approximately 0.75, which indicates a reasonably good prediction performance for a beginner project.
